```python
import gensim
import os
```

```python
import nltk 
nltk.download('punkt')
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
story = []
for filename in os.listdir('/content/drive/MyDrive/Data'):
    
    f = open(os.path.join('/content/drive/MyDrive/Data',filename))
    corpus = f.read()
    raw_sent = sent_tokenize(corpus)
    for sent in raw_sent:
        story.append(simple_preprocess(sent))
```

```python
story
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2,
    workers=4
)
```

```python
model.build_vocab(story)
```

```python
model.train(story,total_examples=model.corpus_count,epochs=model.epochs)
```

```python
model.wv.most_similar('tell')
```

```python
model.wv.most_similar('hysterical')
```

```python
model.wv.doesnt_match(['hysterical','dismiss','dialing','buzzer','ben','Phoebe'])
```

```python
model.wv['hysterical'].shape
```

```python
model.save("word2vec.model")
```
